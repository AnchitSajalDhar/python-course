```python
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn as sk
```

```python

pd1["total_benefitted"] = pd1["No. of Inmates benefitted by Elementary Education"] + pd1["No. of Inmates benefitted by Adult Education"] + pd1["No. of Inmates benefitted by Higher Education"] + pd1["No. of Inmates benefitted by Computer Course"]
pd1.head()
```

```python
pd1 = pd1.append(pd1.sum(), ignore_index=True)
pd1.at[35, "STATE/UT"] = "totals"
pd1.head()
```

```python
x=pd1["STATE/UT"]
y=pd1["total_benefitted"]

plt.bar(x,y)
plt.setp(plt.gca().get_xticklabels(), rotation=90,horizontalalignment='right') 
plt.title("STATE/UT vs Total_inmates_benefitted")
plt.xlabel("STATE/UT")
plt.ylabel("Total_inmates_benefitted")
```

```python
labels = pd1["STATE/UT"]
values = pd1["total_benefitted"]

plt.pie(values, labels = labels,radius=20.5,autopct = "%0.2f%%",textprops = {"fontsize":10},startangle=270,pctdistance = 0.6, shadow=True)
plt.title("STATE/UT vs Total_inmates_benefitted")
```
